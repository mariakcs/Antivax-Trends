```python
vaers_test = [vaers_2000_csv_df, vaers_2001_csv_df, vaers_2002_csv_df, vaers_2003_csv_df, vaers_2004_csv_df, vaers_2005_csv_df, 
                          vaers_2006_csv_df, vaers_2007_csv_df, vaers_2008_csv_df, vaers_2009_csv_df, vaers_2010_csv_df,
                          vaers_2011_csv_df, vaers_2012_csv_df, vaers_2013_csv_df, vaers_2014_csv_df, vaers_2015_csv_df,
                          vaers_2016_csv_df, vaers_2017_csv_df, vaers_2018_csv_df]
vaers2000_2018_dfs = []

for vaers2000_2018_df in vaers_test:
    vaers2000_2018_df = vaers2000_2018_df.drop(columns=['VAERS_ID', 'CAGE_YR', 'CAGE_MO','L_THREAT', 
                                                        'ER_VISIT','HOSPITAL', 'HOSPDAYS', 'X_STAY', 'NUMDAYS', 
                                                        'LAB_DATA', 'V_ADMINBY', 'V_FUNDBY','CUR_ILL', 'SPLTTYPE', 
                                                        'FORM_VERS', 'TODAYS_DATE',  'ER_ED_VISIT'])
    vaers2000_2018_df = vaers2000_2018_df.loc[vaers2000_2018_df["STATE"] == "CA", :]
    vaers2000_2018_df = vaers2000_2018_df[~vaers2000_2018_df['SYMPTOM_TEXT'].isnull()]
    vaers2000_2018_dfs.append(vaers2000_2018_df)
```

```python
vaers2000_2018_df = pd.concat(vaers2000_2018_dfs)
vaers2000_2018_df.head()
```

```python
vaers2000_2018_df = vaers2000_2018_df.dropna(axis='columns')
vaers2000_2018_df.head()
```

```python
vaers_total_symptom = vaers2000_2018_df.groupby("RECVDATE").count()["SYMPTOM_TEXT"]
vaers_total_symptom.index = pd.to_datetime(vaers_total_symptom.index)
vaers_total_symptom_df = pd.DataFrame(vaers_total_symptom)
x_axis = np.arange(len(vaers_total_symptom_df))
plt.scatter(vaers_total_symptom_df.index, vaers_total_symptom_df.values, color='b', alpha=0.25)
plt.xticks(rotation=90)
plt.figure(figsize=(50,140))
plt.show()
```
