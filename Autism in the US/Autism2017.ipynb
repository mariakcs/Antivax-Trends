```python
fips_only=pd.DataFrame(autism_cdc2["FIPSST"].apply(lambda x: str(x).split('b')[1][1:3]))
fips_only
```

```python
autism_selected=autism_cdc2[['FIPSST','TOTKIDS_R','SC_AGE_YEARS','SC_SEX','MOMAGE','K2Q35A','K2Q35C','K2Q35A_1_YEARS','TOTCSHCN','AGEPOS4','K6Q40','BREASTFEDEND_MO_S','BULLIED',"BIRTHWT_VL","K2Q05"]]
autism_selected["FIPS"]=fips_only
autism_selected.head()
```

```python
autism_only_2017=autism_selected[(autism_selected["K2Q35A"]==1)]
autism_only_2017.head()
```

**Cleaning AutismDF and renaming columns**

```python
autism_2017df=autism_only_2017.rename(columns={"TOTKIDS_R":"kids_household",
                                               "SC_AGE_YEARS":"kid_yrs",
                                               "SC_SEX":"kid_gender",
                                               "K2Q35A":"ASD_diagnosed",
                                               "K2Q35C":"ASD_Severity", })
```

```python
autism_2017df["ASD_diagnosed"].count()
```

```python
f, ax = plt.subplots(1, figsize=(16, 10))
autism_bygender=sn.countplot(x="kid_gender",data=autism_2017df,palette="coolwarm")
labels=["Boys","Girls"]
ax.set_xticklabels(labels, fontsize=14)
plt.xlabel('Gender', fontsize=18)
plt.ylabel('Autism cases', fontsize=18)
plt.title("Autism diagnosed cases by Gender in the U.S",fontsize=20)
plt.savefig("Autism diagnosed cases by Gender in the U.S 2017.png")
plt.show
```

```python
f, ax = plt.subplots(1, figsize=(16, 10))
autism_bullied=sn.countplot(x="BULLIED",data=autism_2017df,palette="hls")
labels=["Definitely true","Somewhat true","Not true"]
ax.set_xticklabels(labels, fontsize=14)
plt.xlabel('Bullied', fontsize=18)
plt.ylabel('Number of cases', fontsize=18)
plt.title("Autistic Kids bullied level",fontsize=20)
plt.savefig("Autism Bullied 2017.png")
plt.show
```

```python
f, ax = plt.subplots(1, figsize=(16, 10))
momage_autism=sn.countplot(x="MOMAGE",data=autism_2017df,palette="hls")
plt.xticks(rotation="90")
plt.xlabel('Mother Age', fontsize=18)
plt.ylabel('Number of cases', fontsize=18)
plt.title("Autism cases by mother age",fontsize=20)
plt.savefig("Mother age 2017.png")
plt.show
```

```python
f, ax = plt.subplots(1, figsize=(16, 10))
autism_bullied=sn.countplot(x="BIRTHWT_VL",data=autism_2017df,palette="hls")
labels=["Yes","No"]
ax.set_xticklabels(labels, fontsize=14)
plt.xlabel('Low weight <1500g', fontsize=18)
plt.ylabel('Number of cases', fontsize=18)
plt.title("Autism and birth weight",fontsize=20)
plt.savefig("Autism birth weight 2017.png")
plt.show
```

```python
f, ax = plt.subplots(1, figsize=(16, 10))
autism_bullied=sn.countplot(x="K2Q05",data=autism_2017df,palette="hls")
labels=["Yes","No"]
ax.set_xticklabels(labels, fontsize=14)
plt.xlabel('premature 3 or more weeks before due date', fontsize=18)
plt.ylabel('Number of cases', fontsize=18)
plt.title("Autism related to being premature",fontsize=20)
plt.savefig("Autism premature.png")
plt.show
```

```python
fp="States 21basic/geo_export_d0f8ce85-b142-405a-a64d-dd3928ba0f98.shp"
map_df=gpd.read_file(fp)
map_df.rename(columns={"state_fips":"FIPS"},inplace=True)

map_df.head()
```

```python

# Setup figure and axis with different size
f, ax = plt.subplots(1, figsize=(45, 25))
# Plot layer of polygons on the axis
map_df.plot(ax=ax)
# Remove axis frames
ax.set_axis_off()
# Display
plt.show()
```

```python
autism_by_state=pd.DataFrame(autism_only_2017.groupby("FIPS")["K2Q35A"].count()).rename(columns={"K2Q35A": "Autism_Cases"})
autism_by_state.describe()
```

```python
merged = map_df.join(autism_by_state,on="FIPS",how="right")
merged["state_name"]
```

```python
# set a variable  to visualise on the map
variable = "Autism_Cases"
# set the range for the choropleth
vmin, vmax = 0, 25
# create figure and axes for Matplotlib
fig, ax = plt.subplots(1, figsize=(15, 10))
# create map
merged.plot(cmap="YlOrRd",column=variable, linewidth=0.5, ax=ax, edgecolor="0.8")
# remove the axis
ax.axis("off")
# add a title
ax.set_title("Number of Autism Cases in the US in 2017 per 1,000 children", fontdict={"fontsize": "25", "fontweight" : "4"})
# create an annotation for the data source
ax.annotate("Source: National Survey of Children's health, 2017",xy=(0.8, .08),  xycoords="figure fraction",
            horizontalalignment="right", verticalalignment="top", fontsize=12, color="#555555")
# Create colorbar as a legend
sm = plt.cm.ScalarMappable( cmap="YlOrRd",norm=plt.Normalize(vmin=vmin, vmax=vmax))
# empty array for the data range
sm._A = []
# add the colorbar to the figure
cbar = fig.colorbar(sm)
plt.savefig("Number of Autism Cases in the US in 2017.png")
plt.show
```
